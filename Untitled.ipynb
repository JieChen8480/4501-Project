```python
import bs4
import requests
import pyarrow.parquet as pq
import pyarrow
import pandas as pd
import re

TAXI_URL = "https://www1.nyc.gov/site/tlc/about/tlc-trip-record-data.page"


def get_taxi_html():
    response = requests.get(TAXI_URL)
    html = response.content
    return html


def find_taxi_parquet_links():
    # YOUR CODE HERE
    soup = bs4.BeautifulSoup(get_taxi_html(), 'html.parser')
    soup = soup.find_all('a')
    a = []
    for i in soup:
        if "Yellow Taxi Trip Records" in i :
            a.append(str(i))
    return a
    raise NotImplementedError()
uber_data = pd.read_csv("uber_rides_sample.csv")
weather_data = pd.DataFrame()
for i in range(2009,2015):
    weather_data = weather_data.append(pd.read_csv(f"{i}_weather.csv"))
sample = pd.read_csv("sample.csv")
```

```python
urls = []
pattern= r"https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_\d{4}-\d{2}.parquet"
for i in find_taxi_parquet_links():
    matches = re.search(pattern, i)
    k = matches.group()
    if (k[64:68]) in ['2009','2010','2011','2012','2013','2014']:
        urls.append(matches.group())
    if '2015' in k and k[69:71] in ['01','02','03','04','05','06']:
        urls.append(matches.group())
filelist = []
for i in urls:
    filelist.append(f"taxi_data_{i[64:68]}_{i[69:71]}.parquet")
```

```python
import requests
table = pd.DataFrame()
for i in urls[-12:]:
    response = requests.get(i, stream=True)
    with open(f"taxi_data_{i[64:68]}_{i[69:71]}.parquet", "wb") as f:
        for chunk in response.iter_content(chunk_size=1024): 
            if chunk:
                f.write(chunk)
    table =(pd.read_parquet(f"taxi_data_{i[64:68]}_{i[69:71]}.parquet", engine='pyarrow'))
    if i[64:68] in ["2011","2012","2013","2014","2015"]:
        table = table.drop(['VendorID','tpep_dropoff_datetime','RatecodeID','store_and_fwd_flag','PULocationID','DOLocationID','payment_type','extra','mta_tax','tip_amount','tolls_amount','improvement_surcharge','total_amount','congestion_surcharge','airport_fee'],axis = 1)
        table.rename(columns ={'tpep_pickup_datetime' : 'pickup_datetime'})
    elif i[64:68] == "2009":
        table = table.drop(['Trip_Dropoff_DateTime','vendor_name','Rate_Code','store_and_forward','Payment_Type',"surcharge","mta_tax","Tolls_Amt",'Tip_Amt','Total_Amt'],axis = 1)
        table = table.rename(columns ={'Fare_Amt':'fare_amount','Trip_Pickup_DateTime':'pickup_datetime','Start_Lon':'pickup_longitude','Start_Lat':'pickup_latitude','End_Lon':'dropoff_longitude','End_Lat':'dropoff_latitude','Passenger_Count':'passenger_count','Trip_Distance':'trip_distance'})
    else:
        table = table.drop(['dropoff_datetime','vendor_id','rate_code','store_and_fwd_flag','payment_type',"surcharge","mta_tax","tolls_amount",'tip_amount','total_amount'],axis = 1)
    table.to_parquet(f"taxi_data_{i[64:68]}_{i[69:71]}.parquet")
```

```python
#convert taxi data to sql
import sqlite3
connection = sqlite3.connect("taxi.db")
c = connection.cursor()
    
SQL_STMT = """
CREATE TABLE IF NOT EXISTS taxi
(
    id INTEGER PRIMARY KEY,
    datetime DATETIME,
    fareAmount FLOAT,
    distance FLOAT
)
"""
with connection:
    c.execute(SQL_STMT)
connection.commit()

df = sample
df = pd.DataFrame(df, columns= ['pickup_datetime','fare_amount','trip_distance'])
df.to_sql('taxi', connection, if_exists='replace', index = False)

c.execute('''  
    SELECT * FROM taxi
          ''')

for row in c.fetchall():
    print (row)
```

```python
# create schema
import sqlite3
import logging
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Date, Integer, Float, String, DateTime
from sqlalchemy.orm import validates
from sqlalchemy import create_engine
import datetime

engine = create_engine(f"sqlite:///taxi.db", echo=True)
sqllogger = logging.getLogger("sqlalchemy.engine.Engine")
formatter = logging.Formatter("[sqlalchemy] %(message)s")

Base = declarative_base()

df = sample
df = pd.DataFrame(df, columns= ['pickup_datetime','fare_amount','trip_distance'])

class Taxi(Base):
    __tablename__ = "taxi"  
    id = Column(Integer, primary_key=True)
    pickup_datetime = Column(DateTime, primary_key=True)
    fareAmount = Column(Float)
    distance = Column(Float)

class Uber(Base):
    __tablename__ = "uber"  
    id = Column(Integer, primary_key=True)
    pickup_datetime = Column(DateTime, primary_key=True)
    fareAmount = Column(Float)
    distance = Column(Float)
```

```python
Taxi.__table__
Uber.__table__
Base.metadata.create_all(engine, checkfirst=True)

df_taxi = pd.DataFrame(df, columns= ['pickup_datetime','fare_amount','trip_distance'])
df_taxi.to_sql('taxi', engine, if_exists='replace', index = False)
df_uber = pd.DataFrame(uber_data, columns= ['pickup_datetime','fare_amount','trip_distance'])
df_uber.to_sql('uber', engine, if_exists='replace', index = False)
```
